```python
import uuid, random, matplotlib.pyplot as plt

class Person:
    age: int = 0
    gender: str
    peaceful: bool
    food: int = 0
    def __init__(self, gender: str, peaceful: bool):
        self.gender = gender
        self.peaceful = peaceful
    def get_age(self) -> int:
        return self.age
    def increase_age(self) -> None:
        self.age += 1
    def get_gender(self) -> str:
        return self.gender
    def get_peaceful(self) -> bool:
        return self.peaceful
    def get_food(self) -> int:
        return self.food
    def set_food(self, meal: int) -> None:
        self.food += meal
    def __str__(self) -> str:
        return f"Age: {self.age} | Gender: {self.gender} | Peaceful: {self.peaceful} | Food: {self.food}"
        
```

```python
#Initial Starting Population
STARTING_POPULATION: int = 30
POPULATION: dict[str, Person] = {}
TOTAL_FOOD: int = 50
YEAR_DICT: dict[int, int] = {}

def generate_name() -> str:
    return uuid.uuid4().hex[:6].upper()
```

```python
def initial_pop(pop: int, pop_dict: dict[str, Person]) -> None:
    for i in range(pop):
        pop_dict[generate_name()] = Person(random.choice(["male", "female"]), random.choice([True, False]))

def years(years: int, pop_dict: dict[str, Person]) -> None:
    for i in range(years):
        YEAR_DICT[i] = len(pop_dict)
        full_dict: dict[str, Person] = pop_dict.copy()
        empty_dict: dict[str, Person]
        for i in range(len(list(full_dict))//2):
            person1 = random.choice(list(full_dict.keys()))
            del full_dict[person1]
            person2 = random.choice(list(full_dict.keys()))
            del full_dict[person2]
            if POPULATION[person1].get_peaceful() and POPULATION[person2].get_peaceful():
                empty_dict[person1] = POPULATION[person1]
                empty_dict[person2] = POPULATION[person2]
            else:
                x = random.choice[person1, person2]
                empty_dict[x] = POPULATION[x]
            
                                                                                
            
        
        
        
        
        
            
            
            
            
                
```

```python
initial_pop(STARTING_POPULATION, POPULATION)
years(50, POPULATION)
lists = sorted(YEAR_DICT.items()) # sorted by key, return a list of tuples
x, y = zip(*lists)
print(len(POPULATION))
plt.plot(x,y)
```
